# NLP com spaCy

```python
# Importando spaCy
import spacy

# Carregando modelo
nlp = spacy.load("pt_core_news_lg")
```

```python
# Processando
texto = "Maria trabalha na Microsoft."

doc = nlp(texto)

```

```python
print(type(doc))
print(doc)
```

```python
dados = doc.to_dict()
for k, v in dados.items():
    print(f"Chave: {k:25} | Valor {type(v)}")
```

```python
# Token Por Token
for token in doc:
    print(token)
```

```python
# Token pela Posição
print(doc[0])

# Texto original
print(doc.text)
```

## Atributos dos tokens

```python
token = doc[0]
print(token)

# Verifica composto apenas por letras
print(token.is_alpha)

# Verifica se é um stopword
print(token.is_stop)

# Verifica se é pontuação
print(token.is_punct)
```

```python
# Formato dos tokens (Maiúsculas e munúsculas)
for token in doc:
    print(token, token.shape_)
```

## Separação de Frases

```python
texto = "Hoje fez sol! Amanhã poderá chover. Depois voltará a fazer calor"
doc = nlp(texto)

for frase in doc.sents:
    print(frase)
```

# Lematização
Lematizar signfica encontrar a forma "base" da palavra.

```python
texto = "As crianças estavam correndo felizes."
doc = nlp(texto)
for token in doc:
    print(token, "->", token.lemma_)
```

# Classes gramaticais

```python
texto = "O menino correi rapidamente para casa."
doc = nlp(texto)

for token in doc:
        print(token.text, "->", token.pos_)
```

# Entidades

```python
texto = "João trabalha na microsoft em São Paulo."
doc = nlp(texto)

for entidade in doc.ents:
        print(entidade.text, "->", entidade.label_)
```

# Vetores de Similaridade


```python
nlp = spacy.load("en_core_web_lg")

pairs = [
    ("car", "automobile"),
    ("dog", "cat"),
    ("Brazil", "Argentina"),
    ("Computer", "Keyboard"),
    ("banan", "airplane")

]

for word1, word2 in pairs:
    doc1 = nlp(word1)
    doc2 = nlp(word2)

    print(f"{word1:10} x {word2:10} Semelhança: {doc1.similarity(doc2)*100:.2f}%")

```

# Em português

```python
nlp = spacy.load("pt_core_news_lg")

pairs = [
    ("carro", "volante"),
    ("cão", "gato"),
    ("Brazil", "Argentina"),
    ("Computador", "Notebook"),
    ("banana", "maça")

]

for word1, word2 in pairs:
    doc1 = nlp(word1)
    doc2 = nlp(word2)

    print(f"{word1:10} x {word2:10} Semelhança: {doc1.similarity(doc2)*100:.2f}%")

```
